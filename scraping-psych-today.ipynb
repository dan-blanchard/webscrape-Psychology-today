## Scaping Psychology Today for Number of Mental Health Professionals in NC by City

### Load libraries

```python
print("importing libraries...")

import time

import lxml
from bs4 import BeautifulSoup
from selenium import webdriver
```

## Lets get scraping!

```python
def count_city_therapists(driver, city_url, recs_per_page=20):
    """Returns the number of therapists at the given city URL"""
    current_url = city_url
    while True:
        # Sleep to avoid bot detection
        time.sleep(1)
        driver.get(current_url)
        soup = BeautifulSoup(driver.page_source, 'lxml')
        next_page_links = soup.select('a.btn-default.pager-page')
        has_next_page = soup.select('a.btn-next')
        if has_next_page:
            current_url = next_page_links[-1]['href']
        else:
            pager_here = soup.select('a.pager-page.here')
            final_page_num = int(pager_here[0].text) if pager_here else 1
            current_page_count = len(soup.select('div.normal-result.row'))
            return current_page_count + (final_page_num - 1) * recs_per_page
                                            
        
```

```python
state_url = 'https://www.psychologytoday.com/us/therapists/north-carolina'
driver = webdriver.Chrome()
driver.get(state_url)
soup = BeautifulSoup(driver.page_source, 'lxml')
cities = soup.findAll('a', {'data-event-label': 'RefineCityLink'})
```

```python
for city in cities:
    print(f"{city.text}\t{count_city_therapists(driver, city['href'])}")
```
